## Resumo Teórico

Os testes de hipótese são procedimentos estatísticos usados para avaliar afirmações sobre uma população com base em dados de uma amostra. Primeiro, formulam-se duas hipóteses: a hipótese nula (H₀), que representa a suposição inicial, e a hipótese alternativa (H₁), que representa uma possível diferença ou efeito. Utiliza-se a curva normal e fórmulas de probabilidade para calcular a chance de observar os resultados obtidos. Com base nesse cálculo, decide-se aceitar ou rejeitar H₀, considerando o nível de significância (α) e o valor-p, garantindo assim decisões fundamentadas em evidências estatísticas.

---


## Exemplo Ilustrativo

Suponha que um estudo em determinada região mostra que a ingestão diária média de calórias em adultos é de 2400Kcal. Considere que em um grupo de 25 adultos desta população apresentou um consumo médio de 3000Kcal, com um desvio padrão populacional de 1250Kcal. Para testar se o consumo calórico deste grupo é diferente do padrão de consumoo da população, pode ser efetivado o teste da Normal para médias, como mostrado a seguir

## Dados importantes e suas devidas análises

- Média populacional -> μ = 2400Kcal
- Amostra = 25
- x = 3000
- Desvio Populacional  = 1250Kcal

– Média populacional (μ): é a média verdadeira da população inteira, ou seja, o valor esperado se medíssemos todos os indivíduos (no exemplo, 2400 Kcal).

– Amostra: é um pequeno grupo retirado da população para representar o todo (no exemplo, 25 adultos).

– Média amostral (x̄): é a média calculada com os dados da amostra (no exemplo, 3000 Kcal).

– Desvio padrão populacional (σ): mede quanto os valores da população variam em torno da média, indicando o grau de dispersão (no exemplo, 1250 Kcal).

---

## Passo 1

H₀ (hipótese nula): a média da amostra é igual à média populacional (x̄ = μ).
→ Ou seja, não há diferença entre o grupo e a população.

H₁ (hipótese alternativa): a média da amostra é diferente da média populacional (x̄ ≠ μ).
→ Ou seja, há diferença entre o grupo e a população.

---

## Passo 2

## Cálculo de Z, onde EP é o Erro Padrão da Média. 

O EP ajuda a avaliar a confiabilidade da média amostral calculada.

## Fórmulas Imporantes!



Z = 3000 - 2400 / 250 = 2,4





EP = 1250 / √25 -> 250

---

## Passo 3

## Determinação do parametro p.

O valor da aréa hachurada, é o valor que queremos encontrar!


Aí fazemos os seguintes passos:
- Calculamos o valor da área de -∞ até 2,4. Chamaremos isso de A1.

A1 = 1 - (valor de pertinência de 2,4) -> 1 - 0,9918 = 0,0082

Importante: Para achar o valor de pertinência, basta olhar a tabela de distribuição de probabilidade.

- Agora calculamos o valor de p:

p = 2 x (1 - 0,9918) = 0.0164 = 1,64%

---

## Passo 4

## Tomada de Decisão.

- Para a tomada de decisão, olharemos na tabela abaixo, o nível se significância de p, e a partir disso, faremos a nossa decisão de aceitar ou rejeitar H0. 

No nosso caso, o H0 será rejeitado, pois o p é igual a 1,64, ou seja, está entre 1% e 5%, com isso, há maior certeza para rejeitar H0.

---

## Passo 5

## Conclusão.

- Baseado no experimento que foi realizado, devemos fazer a conclusão do que foi obtido!

No nosso caso, a conclusão ideal seria: "Concluí-se que o consumo da amostra pode ser considerado diferente do da população, a um nível de significância de 1,64%".

## Questão1

Suponha que um estudo mostra que a média de horas de sono de adultos em uma cidade é de 7 horas por noite. Um grupo de 36 adultos dessa população apresentou uma média de 6 horas de sono, com um desvio padrão populacional de 1,8 hora. Para verificar se esse grupo dorme de forma diferente da média da população, pode-se aplicar o teste da Normal para médias.

- Faça os cálculos e chegue à conclusão do resultado obtido.



```python
# Média populacional = u
# Média amostral = x
# Desvio padrão populacional = o
# Desvio padrão amostral = s
# Tamanho da amostra = n
# Nível de significância = a
# Nível de confiança = (1 - a) . 100%
u = 7
x = 6
n = 36
o = 1.8
```

```python
ep = o / (n ** 0.5)
z = (x - u) / ep
print(f"z = {abs(z)}")
#Valor de pertinência = vp
vp = 4.342 * 10**(-4) # Valor obtido pela tabela usada em Probabilidade e Processos Estocásticos do Inatel, a qual já dá o valor diretamente
print(f"Valor de pertinência = {vp}")
p = 2*vp*100
print(f"p-valor = {p}%")
```

```python
if p <= 1:
    print("Certeza máxima para rejeitar H0")
elif p <= 5:
    print("Certeza grande para rejeitar H0")
elif p <= 10:
    print("Certeza grande para aceitar H0")
else:
    print("Certeza máximo para aceitar H0")
```

Conclui-se que as horas de sono da amostra pode ser considerado diferente do da população, a um nível de significância de 0,0868%.
