# ATIVIDADE 12 - Tabela de frequências e gráficos para variáveis qualitativas.

**Objetivo:** Reforçar a construção de tabelas de frequência absoluta, relativa e acumulada.

---

Uma pesquisa foi feita com 50 estudantes sobre o principal meio de transporte usado para ir até a escola. As respostas obtidas foram:

| Meio de transporte | Quantidade de estudantes |
| ------------------ | ------------------------ |
| Carro              | 12                       |
| Ônibus             | 20                       |
| Bicicleta          | 8                        |
| A pé               | 10                       |


Questão 1 - Monte uma tabela de frequências (Frequência Relativa e Frequência Acumulada)

```python
import pandas as pd

# Utilize o DataFrame
df = pd.DataFrame({
    'Meio de Transporte': ['Carro', 'Ônibus', 'Bicicleta', 'A pé'],
    'Quantidade de Estudantes': [12, 20, 8, 10],
    'Frequência Relativa': [12/50, 20/50, 8/50, 10/50],
    'Frequência Acumulada': [12, 32, 40, 50]
})

df
```

Questão 2 - Crie um gráfico de barras

```python
import matplotlib.pyplot as plt
x = df['Meio de Transporte']
y = df['Quantidade de Estudantes']
plt.title('Quantidade de Estudantes por Meio de Transporte')
plt.xlabel('Meio de Transporte')
plt.ylabel('Quantidade de Estudantes')
plt.xticks(rotation=45)
plt.bar(x, y)
plt.show()
```

Questão 3 - Crie um gráfico de pizza

```python
plt.figure(figsize=(8, 8))
plt.pie(df['Frequência Relativa'], labels=df['Meio de Transporte'], autopct='%1.1f%%', startangle=140)
plt.title('Distribuição de Estudantes por Meio de Transporte')
plt.axis('equal')
plt.show()
```

Questão 4 - Interprete os resultados
- Qual meio de transporte é mais utilizado?
- Qual meio de transporte é menos utilizado?
- Qual é a diferença percentual entre o mais e o menos frequente?

```python
# Qual meio de transporte é mais utilizado
mais_utilizado = df.loc[df['Frequência Relativa'].idxmax(), 'Meio de Transporte']
print(f"O mais utilizado é: {mais_utilizado}")

# Qual meio de transporte menos utilizado
menos_utilizado = df.loc[df['Frequência Relativa'].idxmin(), 'Meio de Transporte']
print(f"O menos utilizado é: {menos_utilizado}")

# Qual é a diferença percentual entre o mais e o menos frequente
diferenca_percentual = df.loc[df['Frequência Relativa'].idxmax(), 'Frequência Relativa'] - df.loc[df['Frequência Relativa'].idxmin(), 'Frequência Relativa']
print(f"A diferença percentual é: {diferenca_percentual:.2%}")
```
